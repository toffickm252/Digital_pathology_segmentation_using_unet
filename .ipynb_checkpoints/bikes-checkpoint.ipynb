### Data Preprocessing

```python
X=df.drop(['is_equity','ride_id','started_at','ended_at','start_station_id','end_station_id','member_casual','rideable_type','start_station_name','end_station_name'],axis=1)
```

```python
Y=df.is_equity
```

```python
## Label Encoding
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
en=LabelEncoder()
```

```python
Y=en.fit_transform(Y)
```

### Feature Standardization

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler=MinMaxScaler((-1,1))
features=scaler.fit_transform(X)
```

```python
target=Y
```

### Feature Selection

```python
from sklearn.feature_selection import RFE
from sklearn.linear_model import LogisticRegression
```

```python
model = LogisticRegression()
rfe = RFE(model, 5)
fitted = rfe.fit(features,target)
```

```python
print("Num Features:", fitted.n_features_)
print("Selected Features:", fitted.support_)
print("Feature Ranking:", fitted.ranking_)
```
